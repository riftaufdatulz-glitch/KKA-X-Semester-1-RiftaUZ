#Aplikasi menghitung Luas Persegi Panjang

```python
panjang = int(input("Masukkan panjang dari persegi panjang (cm): "))
lebar = int(input("Masukkan lebar dari persegi panjang (cm): "))

luas = panjang * lebar

print(f"\nLuas dari persegi panjangnya adalah {luas} cm^2.")

```

#Aplikasi Menghitung Volume Bangun Balok

```python
panjang = int(input("Masukkan panjang dari balok (cm): "))
lebar = int(input("Masukkan lebar dari balok (cm) "))
tinggi = int(input("Masukkan tinggi darai balok (cm) "))

volume = panjang * lebar * tinggi

print(f"\nVolume dari baloknya adalah {volume} cm^2.")
```
